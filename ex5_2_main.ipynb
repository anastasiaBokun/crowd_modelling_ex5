```python
from linear_vector_field_estimator import LinearVectorFieldEstimator
import numpy as np

def main():
    """
    Main function to estimate the linear vector field, solve the linear system, and visualize the results.
    """
    # Create LinearVectorFieldEstimator instance
    estimator = LinearVectorFieldEstimator('linear_vectorfield_data_x0.txt', 'linear_vectorfield_data_x1.txt')

    # Estimate the matrix A
    estimator.estimate_matrix_A()
    print(f"Estimated matrix A_hat:\n {estimator.A_hat}")

    # Solve the linear system with initial point x0 and time T
    T = estimator.dt
    x0 = np.zeros(2)
    solution = estimator.solve_linear_system(T, x0)

    # Compute mean squared error
    mse = estimator.compute_mean_squared_error(solution)
    print(f"Mean Squared Error: {mse}")

    # Solve the extended linear system with initial point x0_extended and end time Tend
    Tend = 100
    x0_extended = np.array([10, 10])
    solution_extended = estimator.solve_extended_system(Tend, x0_extended)

    # Plot the trajectory
    estimator.plot_trajectory(solution_extended)

    # Plot the phase portrait
    estimator.plot_phase_portrait(solution_extended)


if __name__ == '__main__':
    main()
```
